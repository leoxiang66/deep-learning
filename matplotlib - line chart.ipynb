````python
import matplotlib.pyplot as plt 
import numpy as np

n=24
y1 =np . random. randint (27,37, n)
y2 =np . random. randint (40, 60, n)
plt.plot(y1, label='Temperature')
plt.plot(y2, label='Humidity')
plt. xlim (0, 23)
plt. ylim (20, 70)
plt.xlabel( 'hour', fontsize=12)
plt.ylabel ( 'measure value',fontsize = 12)
plt.title('24 hour temperature and humidity measurement', fontsize=16)
plt. legend () 
plt. show ()
````
