````python
import matplotlib.pyplot as plt 
import numpy as np

# data points
n=1024
x=np. random. normal (0, 1, n)
y=np. random.normal (0, 1, n)
plt. scatter(x, y, color="blue", marker= '*' )

# title
plt.title("normal distribution", fontsize=20) 

# text
plt.text(2.5,2.5, "mean:0\nvariance: 1" )

# axes setup
plt. xlim(-4,4)
plt. ylim(-4,4)
plt.xlabel ( 'X', fontsize=14)
plt.ylabel('y', fontsize=14)

plt. show()
````

````python
import matplotlib.pyplot as plt 
import numpy as np

# data points
n=1024
x1=np. random. normal (0, 1, n)
y1=np. random.normal (0, 1, n)
plt. scatter(x1, y1, color="blue", marker= '*' ,label = 'Normal Distribution')

x2 = np. random. uniform (-4,4,(1, n))
y2 = np. random. uniform (-4,4,(1, n))
plt. scatter(x2, y2, color="yellow", marker= '*' ,label = 'Normal Distribution')



# title
plt.title("Distributions", fontsize=20) 

# text
plt.text(2.5,2.5, "mean:0\nvariance: 1" )

# axes setup
plt. xlim(-4,4)
plt. ylim(-4,4)
plt.xlabel ( 'X', fontsize=14)
plt.ylabel('y', fontsize=14)

plt. show()
````
