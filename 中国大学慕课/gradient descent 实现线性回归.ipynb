### Numpy 实现

1) 加载数据

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np. array([137.97,  104.50,  100.00,  124.32,  79.20,  99.00,  124.00, 
               114.00,106.69,138.05,53.75,46.91,68.00,63.02,81.26,86.21])
y = np. array([145.00,110.00,93.00,116.00,65.32,104.00,118.00,91.00,
               62.00,133.00,51.00,45.00,78.50,  69.65,75.69,  95.30])
```

2) 设置hyperparameter

```python
learn_rate = 0.00001
iter = 100

display_step = 10
```

3) 设置模型参数初始值

```python
np.random.seed(612)
w = np.random.randn()
b = np.random.randn()
```

4) 训练模型



```python
mse = []

for i in range(0,iter+1):
    dl_dw = np.mean(x*(w*x+b-y))
    dl_db = np.mean(w*x+b-y)
    
    w = w - learn_rate * dl_dw
    b = b - learn_rate * dl_db
    
    pred = w*x+b
    Loss = np.mean(np.square(y-pred))/2
    mse.append(Loss)
    
    if i % display_step == 0:
        print(f'i: {i}, Loss: {Loss}, w: {w}, b: {b}')
```

### 结果可视化

```python
plt.figure()

plt.scatter(x,y,color = 'red', label = 'Ground Truth')
plt.scatter(x,pred,color = 'blue', label='Prediction')
plt.plot(x,pred,color = 'blue')

plt.xlabel('Area',fontsize=14)
plt.ylabel('Price',fontsize=14)

plt.legend()
plt.show()
```
