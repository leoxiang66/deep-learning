# Figure object 画布

```python
plt.figure( num,figsize, dpi,facecolor,edgecolor,frameon)
```


- num : 图形编号或名称 , 取值为数字/字符串。
- figsize : 绘图对象的宽和高，单位为英寸。 
- dpi : 绘图对象的分辨率，缺省值为80。 
- facecolor : 背景颜色。 
- edgecolor : 边框颜色。 
- frameon : 表示是否显示边框。




````python
import matplotlib.pyplot as plt
plt.figure(figsize =(3,2), facecolor="green") 
plt.plot() 
plt.show()
````

### 添加标题

1. 添加全局标题

   ```python
   plt.suptitle (标题文字)
   ```

   - suptitle()函数的主要参数 :

2. 添加子标题

	```python
	plt.title(标题文字)
	```
    
    - title()函数的主要参数 :

      <img src="https://i.loli.net/2021/03/17/1hXixeGwSnfEspc.png" alt="image-20210317141129355" style="zoom:50%;" />




````python
import matplotlib.pyplot as plt
fig = plt.figure(facecolor="lightgrey")

plt. subplot (2,2,1)
plt.title( 'subtitle 1') 

plt. subplot (2,2,2)
plt.title( 'subtitle 2 ', loc= "left", color="b")

plt. subplot (2,2,3)
myfontdict ={"fontsize":12, "color": "g", "rotation":30}
plt.title( 'subtitle 3', fontdict=myfontdict)

plt. subplot (2,2,4) 
plt.title( 'subtitle 4', color= 'white', backgroundcolor="black")

plt. suptitle("TITLE", fontsize=20, color="red", backgroundcolor="yellow")

plt. show()
````

3. tight_layout()函数

   检查坐标轴标签、刻度标签、和子图标题，**自动调整子图**，使之填充整个绘图区域，并消除子图之间的重叠。
   ```python
   plt.tight_layout( rect=[left, bottom, right, top])
   ```
   
   <img src="https://i.loli.net/2021/03/17/nDgiOBVep85zauR.png" alt="image-20210317142154224" style="zoom:50%;" />

````python
import matplotlib.pyplot as plt
fig = plt.figure(facecolor="lightgrey")

plt. subplot (2,2,1)
plt.title( 'subtitle 1') 

plt. subplot (2,2,2)
plt.title( 'subtitle 2 ', loc= "left", color="b")

plt. subplot (2,2,3)
myfontdict ={"fontsize":12, "color": "g", "rotation":30}
plt.title( 'subtitle 3', fontdict=myfontdict)

plt. subplot (2,2,4) 
plt.title( 'subtitle 4', color= 'white', backgroundcolor="black")

plt. suptitle("TITLE", fontsize=20, color="red", backgroundcolor="yellow")

plt.tight_layout()

plt. show()
````
